```python
import numpy as np
import cv2
from tensorflow.keras.models import model_from_json,load_model
```

```python
# load the cascade
# cascadeClassifier is the face classification model
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
# capture video from webcam
# cap = cv2.VideoCapture(0)
cap = cv2.VideoCapture('/home/pradeep/mySpace/DAI/projects/music-recommendation-using-face-emotion/data/expressions.mp4')
# to use video file as input
# cap = cv2.VideoCapture('filename.mp4')
```

```python
from deepface import DeepFace
```

```python
import copy

while True:
    # Read the frame/image from the video
    ret, frame = cap.read()
    print(frame)
    img = copy.deepcopy(frame)
    # img = cv2.resize(img,(600,500))
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)
    for (x,y,w,h) in faces:
        fc = gray[y:y+h, x:x+w]
        
        roi = cv2.resize(fc, (48,48))

        result = DeepFace.analyze(img,actions = ['emotion'],enforce_detection=False)
        
        text = result['dominant_emotion']
        cv2.putText(img, text, (x, y-5),
        cv2.FONT_HERSHEY_SIMPLEX, 0.45, (255, 0, 255), 2)
        img = cv2.rectangle(img, (x,y), (x+w, y+h), (0,0,255), 2)
            
    
    cv2.imshow("frame", img)
    key = cv2.waitKey(1) & 0xff
    if key == 27:
        break
    
cap.release()
cv2.destroyAllWindows()
```
